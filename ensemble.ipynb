```python
import pandas as pd
import numpy as np
import random
import os

def reset_seeds(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)


DATA_PATH = "/content/drive/MyDrive/코랩파일들/제주 특산물 가격예측 AI/data/"
SEED = 42

reset_seeds(SEED)
```

```python
submission = pd.read_csv(f'{DATA_PATH}sample_submission.csv')
nlinear = pd.read_csv(f'{DATA_PATH}N_linear.csv')
autogluon = pd.read_csv(f'{DATA_PATH}autogluon.csv')
submission['answer'] = (nlinear['answer'] + autogluon['answer'])/2
```

```python
submission.to_csv('ensemble.csv', index=False)
```
